2..2. Presentation: Object-Oriented Programming (OOP) Concepts in Python
ans:What is OOP?

A programming paradigm based on “objects”.
Helps in organizing code using classes and objects.
Makes code reusable, modular, and easier to maintain.

Key OOP Concepts in Python

Class
Object
Encapsulation
Inheritance
Polymorphism
Abstraction

Class and Object

Class: A blueprint for creating objects
Object: An instance of a class
ex:class Animal:
    def speak(self):
        print("Animal sound")

dog = Animal()  # object
dog.speak()

Encapsulation

Binding data (variables) and methods together
Protects object’s internal state
ex:class Person:
    def __init__(self, name):
        self.__name = name  # private variable

 Inheritance

A class can inherit properties/methods of another class
Promotes code reuse
ex:class Animal:
    def speak(self):
        print("Sound")

class Dog(Animal):
    pass

Polymorphism

Same function name, different behaviors
Ex:class Cat:
    def sound(self):
        print("Meow")

class Dog:
    def sound(self):
        print("Bark")

Abstraction

Hiding internal details, showing only essentials
ex:from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

Why Use OOP in Python?

Cleaner code structure
Reusability
Easier to debug and maintain
Real-world problem modeling

```python
3.
sentence = input("Enter a sentence: ")

sentence = sentence.lower()

words = sentence.split()

word_count = {}

for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1

for word, count in word_count.items():
    print(f"'{word}' occurs {count} time(s)")
```

```python
5.
def add_suffix(s):
    if len(s) < 3:
        return s 
    elif s.endswith('ing'):
        return s + 'ly'  
    else:
        return s + 'ing'
    
string = input("Enter a string: ")

result = add_suffix(string)
print("Modified string:", result)
```

```python
6.
def replace_not_poor(s):
    not_index = s.find('not')
    poor_index = s.find('poor')

    if not_index != -1 and poor_index != -1 and poor_index > not_index:
        s = s[:not_index] + 'good' + s[poor_index + 4:]
    return s

string = input("Enter a sentence: ")

result = replace_not_poor(string)
print("Modified string:", result)
```

```python
7.
def find_gcd(a, b):
    while b:
        a, b = b, a % b
    return a

num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

gcd = find_gcd(num1, num2)
print(f"The GCD of {num1} and {num2} is: {gcd}")
```

```python
17.
import math

def odd_series(n):
    return sum((i*i) / math.factorial(i) for i in range(1, 2*n, 2))

def even_series(n):
    return sum((i*i) / math.factorial(i) for i in range(2, 2*n+1, 2))

n = int(input("Enter number of terms: "))
print("Odd series sum:", odd_series(n))
print("Even series sum:", even_series(n))
```

```python
18.
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n - 1)

num = int(input("Enter a number: "))
print("Factorial:", factorial(num))
```

```python
20.
import random
import string

class User:
    def __init__(self, user_id, name, password):
        self.details = (user_id, name, password) 

    def show_details(self):
        print(f"\nUser Details:\nUser ID: {self.details[0]}\nName: {self.details[1]}\nPassword: {self.details[2]}")

def generate_password(user_input):
    try:
        if not user_input.strip():
            raise ValueError("Input cannot be empty.")

        words = user_input.split()

        chosen_words = ''.join(random.sample(words, min(2, len(words))))
        
        numbers = ''.join(random.choices(string.digits, k=2))
        symbols = ''.join(random.choices('!@#$%^&*', k=2))
        upper = chosen_words.capitalize()

        password = upper + numbers + symbols

        while len(password) < 8:
            password += random.choice(string.ascii_letters + string.digits)

        return password

    except ValueError as ve:
        print("Error:", ve)
        return None
    except Exception as e:
        print("Unexpected error:", e)
        return None

# Main program
try:
    user_id = input("Enter User ID: ")
    name = input("Enter Name: ")
    user_input = input("Enter some words to base your password on: ")

    password = generate_password(user_input)

    if password:
        user = User(user_id, name, password)
        user.show_details()
    else:
        print("Password could not be generated. Please try again.")

except Exception as e:
    print("An error occurred:", e)
```
